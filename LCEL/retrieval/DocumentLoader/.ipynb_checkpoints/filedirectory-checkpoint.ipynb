### File Directory

```python
from langchain.document_loaders import DirectoryLoader
```

```python
loader = DirectoryLoader('C:/Users/Owner/Downloads', glob="**/*.xlsx")
```

```python
docs = loader.load()
```

```python
len(docs)
```

## Show a progress bar

```python
loader = DirectoryLoader('../', glob="**/*.md", show_progress=True)
docs = loader.load()
```

## Use multithreading

```python
loader = DirectoryLoader('../', glob="**/*.md", use_multithreading=True)
docs = loader.load()
```

## Change loader class

```python
from langchain.document_loaders import TextLoader
```

```python
loader = DirectoryLoader('../', glob="**/*.md", loader_cls=TextLoader)
docs = loader.load()
len(docs)
```

## Python Loader
## Silent fail (silent_errors=True)

```python
from langchain.document_loaders import PythonLoader
```

```python
loader = DirectoryLoader('../../../../../', glob="**/*.py", loader_cls=PythonLoader,silent_errors=True)
```

```python
docs = loader.load()
len(docs)
```

```python
doc_sources = [doc.metadata['source']  for doc in docs]
doc_sources
```

## Auto detect encodings

```python
path = '../../../../../tests/integration_tests/examples'
text_loader_kwargs={'autodetect_encoding': True}
loader = DirectoryLoader(path, glob="**/*.txt", loader_cls=TextLoader, loader_kwargs=text_loader_kwargs)
docs = loader.load()
```
